```python
import pandas as pd
from sklearn.cluster import KMeans

# Your dataset
data_gates = {
    'Gate': ['AND2', 'OR2', 'NAND2', 'NOR2', 'XOR2', 'NOT', 'AND3', 'OR3', 'NAND3', 'NOR3'],
    'Delay_ps': [25, 22, 15, 18, 40, 10, 35, 30, 20, 25],
    'Power_uW': [15, 12, 18, 16, 25, 5, 22, 18, 25, 22],
    'Area_um2': [10, 9, 8, 7, 15, 5, 14, 13, 12, 11]
}
df_gates = pd.DataFrame(data_gates)

# Select features for clustering
X_gates = df_gates[['Delay_ps', 'Power_uW', 'Area_um2']]

# Choose the number of clusters
n_clusters_gates = 3

# Initialize and fit the K-means model
kmeans_gates = KMeans(n_clusters=n_clusters_gates, n_init=10)
df_gates['Cluster'] = kmeans_gates.fit_predict(X_gates)

# Analyze the clusters: Mean values
clustered_gates_mean = df_gates.groupby('Cluster').agg({'Delay_ps': 'mean', 'Power_uW': 'mean', 'Area_um2': 'mean', 'Gate': 'count'}).reset_index()
clustered_gates_mean.columns = ['K', 'Avg_Delay_ps', 'Avg_Power_uW', 'Avg_Area_um2', 'Count']
print("\nMean values per cluster:")
print(clustered_gates_mean.to_string(index=False))

# Analyze the clusters: Range of values
clustered_gates_range = df_gates.groupby('Cluster').agg({'Delay_ps': ['min', 'max'], 'Power_uW': ['min', 'max'], 'Area_um2': ['min', 'max']})
clustered_gates_range.columns = ['Min_Delay', 'Max_Delay', 'Min_Power', 'Max_Power', 'Min_Area', 'Max_Area']
print("\nRange of values per cluster:")
print(clustered_gates_range)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Your dataset
data_gates = {
    'Gate': ['AND2', 'OR2', 'NAND2', 'NOR2', 'XOR2', 'NOT', 'AND3', 'OR3', 'NAND3', 'NOR3'],
    'Delay_ps': [25, 22, 15, 18, 40, 10, 35, 30, 20, 25],
    'Power_uW': [15, 12, 18, 16, 25, 5, 22, 18, 25, 22],
    'Area_um2': [10, 9, 8, 7, 15, 5, 14, 13, 12, 11]
}
df_gates = pd.DataFrame(data_gates)

# Select features for clustering
X_gates = df_gates[['Delay_ps', 'Power_uW', 'Area_um2']]

# Choose the number of clusters
n_clusters_gates = 3

# Initialize and fit the K-means model
kmeans_gates = KMeans(n_clusters=n_clusters_gates, n_init=10)
df_gates['Cluster'] = kmeans_gates.fit_predict(X_gates)

# Create a scatter plot of Delay vs. Power, colored by cluster
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df_gates['Delay_ps'], df_gates['Power_uW'], c=df_gates['Cluster'], cmap='viridis', edgecolor='k', s=100)
plt.title('K-means Clustering of Logic Gates (Delay vs. Power)')
plt.xlabel('Delay (ps)')
plt.ylabel('Power Consumption (uW)')
plt.colorbar(scatter, label='Cluster')
plt.grid(True)

# Annotate the points with the gate names (optional, but helpful for identification)
for i, row in df_gates.iterrows():
    plt.annotate(row['Gate'], (row['Delay_ps'], row['Power_uW']), textcoords="offset points", xytext=(5,5), ha='left')

plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Your dataset
data_gates = {
    'Gate': ['AND2', 'OR2', 'NAND2', 'NOR2', 'XOR2', 'NOT', 'AND3', 'OR3', 'NAND3', 'NOR3'],
    'Delay_ps': [25, 22, 15, 18, 40, 10, 35, 30, 20, 25],
    'Power_uW': [15, 12, 18, 16, 25, 5, 22, 18, 25, 22],
    'Area_um2': [10, 9, 8, 7, 15, 5, 14, 13, 12, 11]
}
df_gates = pd.DataFrame(data_gates)

# Select features for clustering
X_gates = df_gates[['Delay_ps', 'Power_uW', 'Area_um2']]

# Scale the features
scaler = StandardScaler()
X_gates_scaled = scaler.fit_transform(X_gates)

# Elbow Method for Optimal 'k'
inertia = []
for k in range(2, 8):  # Try a range of k values
    kmeans = KMeans(n_clusters=k, n_init='auto', random_state=42)
    kmeans.fit(X_gates_scaled)  # Use the scaled data
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 8), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```
